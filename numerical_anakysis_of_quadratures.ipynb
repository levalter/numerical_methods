```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Q(x): # вычисление полинома Q
    return 2 * x + 1
def P5(x): # вычисление полинома P5
    return 480 * x * (8 * x**4 + 40 * x**2 * (x**2 - 1) + 15 * (x**2 - 1)**2)
def P1(x): # вычисление полинома P1
    return 2 * x**2
def f(x, n): # подынтегральная функция
    if n == 1:
        p = P1(x)
    elif n == 5:
        p = P5(x)
    return Q(x) * p

def trapez(N, n): # реализация метода трапеций
    I = 0
    a = -1
    b = 1
    h = (b - a) / N
    for i in range(1, N):
        I += 2 * f(a + i*h, n)
    I += (f(a, n) + f(b, n))
    I *= (h/2)
    return I


def parabol(N, n): # реализация метода парабол
    I = 0
    a = -1
    b = 1
    h = (b - a) / N
    for i in range(1, N):
        I += 2**(1 + i%2) * f(a + i*h, n)
    I += (f(a, n) + f(b, n))
    I *= h/3
    return I


N1 = [10, 11, 20, 21, 30, 31, 40, 41] # сетка для метода трапеций
N2 = [2,4,6, 10, 20, 30, 40] # сетка для метода парабол

print("Метод трапеций, n = 1")
tr1 = []
for i in N1:
    elem = trapez(i, 1)
    tr1.append(elem)
    print("N = ", i, ": ", "%.5f"%elem, sep='') # вызываем метод для каждого значения сетки (из возможных)
print()
print("Метод парабол, n = 1")
par1 = []
for i in N2:
    elem = parabol(i, 1)
    par1.append(elem)
    print("N = ", i, ": ", "%.5f"%elem, sep='')
print()
print("Метод трапеций, n = 5")
tr5 = []
for i in N1:
    elem = trapez(i, 5)
    tr5.append(elem)
    print("N = ", i, ": ", "%.5f"%elem, sep='') # вызываем метод для каждого значения сетки (из возможных)
print()
print("Метод парабол, n = 5")
par5 =[]
for i in N2:
    elem = parabol(i, 5)
    par5.append(elem)
    print("N = ", i, ": ", "%.5f"%elem, sep='')
```

```python
tr1 = np.array(tr1)
tr5 = np.array(tr5)
par1 = np.array(par1)
par5 = np.array(par5)
```

```python
plt.title("Метод трапеций и метод парабол при n=5")
plt.xlabel('Сетка N')
plt.ylabel('Method score')
plt.scatter(N1,tr5,color='gold',label='trapez')
plt.scatter(N2,par5,color='maroon',label='parabol')
plt.legend()
plt.show()
```
